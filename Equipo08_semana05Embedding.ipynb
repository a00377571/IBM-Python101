# **Pregunta - 2:**

Proceso de limpieza. Aplica el proceso de limpieza que consideres adecuado.











```python

# ******* Inlcuye a continuación todas las líneas de código y celdas que requieras: ***********

#Xclean = (
#    df["text"]
#    .str.lower()
#    .str.replace(r'[^a-z\s]+', '', regex=True)
#    .str.replace(r'\s+', ' ', regex=True)
#    .str.strip()
#)
#y = pd.to_numeric(df["label"], errors="raise").astype(int)

y = pd.to_numeric(df["label"], errors="raise").astype(int)

#X = df.review     # Serie de strings
#y = df.label      # Serie de enteros 0s y 1s

negwords = [ 'no', 'nor', 'not', 'ain', 'aren', "aren't", 'don', "don't", 'couldn', "couldn't", 'didn', "didn't", 'doesn', "doesn't", 'hadn', "hadn't", 'hasn', "hasn't", 'haven', "haven't", 'isn', "isn't", 'mightn', "mightn't", 'mustn', "mustn't", 'needn', "needn't", 'shan', "shan't", 'shouldn', "shouldn't", 'wasn', "wasn't", 'weren', "weren't", 'won', "won't", 'wouldn', "wouldn't"]

# Y las excluimos de las stopwords:
mystopwords = [ w for w in stopwords.words('english') if w not in negwords]

def clean_tok(doc):

  #1 convertir a texto los números para no eliminar los registros
  current_data = str(doc)
  #convert to words
  current_data = re.sub(r'(\d+)/(\d+)', lambda m: f"{num2words(int(m.group(1)))} out of {num2words(int(m.group(2)))}",current_data)
  current_data = re.sub(r'\d+', lambda m: num2words(int(m.group(0))),current_data)

  #convertir a minúsculas
  current_data = current_data.lower()

  #convertir emojis a palabras
  current_data = emoji.demojize(current_data, delimiters=(" ", " "))
  current_data = current_data.replace("_"," ")

  #solo letras sin signos de puntuación ni espacios
  pattern_alphabetic =  r'[^a-zA-Z]'
  pattern_extra_spaces = r'\s+'
  current_data = re.sub(pattern_alphabetic, ' ', current_data)
  current_data = re.sub(pattern_extra_spaces, ' ', current_data).strip()

  #tokenización
  tokens = current_data.split()

  #remover stop words
  tokens = [word for word in tokens if word not in mystopwords]

  #token mayor a 1
  tokens = [word for word in tokens if len(word) > 1]

  return tokens

#Xclean = [" ".join(clean_tok(x)) for x in df["text"]]
Xclean = pd.Series([" ".join(clean_tok(x)) for x in df["text"]])


# *********** Aquí termina la sección de agregar código *************
```

```python
def generar_embeddings(model_name, X):
    """
    Regresa los embeddings en un diccionario con las siguientes palabra-embeddings:
    {
        "token": embedding
    }
    """
    model = SentenceTransformer(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    tokens_unicos = set()
    for text in X:
        tokens = tokenizer.tokenize(text)
        for token in tokens:
            tokens_unicos.add(token)
    embeddings = model.encode(list(tokens_unicos))
    embedding_dict = {token: embedding for token, embedding in zip(tokens_unicos, embeddings)}
    return embedding_dict, model
```

# **Pregunta - 7:**




Generamos los vectores embebidos a partir de los conjuntos de entrenamiento, validación y prueba y con las características indicadas en el archivo PDF.

Los llamaremos trainEmb, valEmb y testEmb, respectivamente.


```python
def generar_promedio_embeddings(embedding_dict, X):
    """
    Regresa un df con el promedio de los embeddings de cada texto en X.
    {
        "texto": embedding_promedio
    }
    """
    # not a dict because it eats repeated values, so we will use a list of tuples instead
    embedding_dim = len(next(iter(embedding_dict.values())))
    promedio_embeddings = []
    for text in X:
        tokens = text.split()
        embeddings = [embedding_dict[token] for token in tokens if token in embedding_dict]
        if embeddings:
            promedio_embeddings.append((text, sum(embeddings) / len(embeddings)))
        else:
            promedio_embeddings.append((text, np.zeros(embedding_dim)))
    return promedio_embeddings
```

# **Pregunta - 9:**



Entrenamiento y reporte de los modelos de Regresión Logística y Bosque Aleatorio (Random Forest).


```python
def log_regression(X_t, y_t, X_v, y_v, seed=42):
    """
    Entrena un modelo de regresión logística y evalúa su desempeño en el conjunto de validación.
    Usamos parameter grid search para encontrar los mejores hiperparámetros y reportamos el accuracy obtenido en el conjunto de validación.
    return best_params, best_score, best_model
    """
    # Xtrain comes with the format [(text, embedding)], so we need to extract the embeddings and convert them to a numpy array
    X_train_embeddings = np.array([entry[1] for entry in X_t])
    X_val_embeddings = np.array([entry[1] for entry in X_v])
    param_grid = {
        'C': [0.01, 100],
        'solver': ['liblinear']
    }
    best_score = 0
    best_params = None
    best_model = None
    for params in ParameterGrid(param_grid):
        model = LogisticRegression(**params, random_state=seed)
        model.fit(X_train_embeddings, y_t)
        y_pred = model.predict(X_val_embeddings)
        score = accuracy_score(y_v, y_pred)
        if score > best_score:
            best_score = score
            best_params = params
            best_model = model
    print(f"Mejores hiperparámetros: {best_params}")
    print(f"Accuracy en el conjunto de validación: {best_score:.4f}")
    print(f"Accuracy en el conjunto de entrenamiento: {accuracy_score(y_t, best_model.predict(X_train_embeddings)):.4f}")
    return best_params, best_score, best_model
```

```python

def random_forest(X_train, y_train, X_val, y_val, seed=42):
    """
    Entrena un modelo de Random Forest optimizado mediante Grid Search manual.
    Usa procesamiento en paralelo para reducir drásticamente el tiempo de ejecución.
    """
    # 1. Extraer embeddings eficientemente
    X_train_embeddings = np.array([embedding for text, embedding in X_train])
    X_val_embeddings = np.array([embedding for text, embedding in X_val])

    param_grid = {
        'n_estimators': [200, 300],
        'max_depth': [10, 14],
        'min_samples_split': [5, 10, 20],
        'min_samples_leaf': [2, 4, 8],
        'max_features': ['log2']
    }

    best_score = 0
    best_params = None
    best_model = None

    # Lista para almacenar los resultados y evitar re-entrenar el modelo al final
    historial_resultados = []

    print("Iniciando Grid Search optimizado...")

    # UN SOLO BUCLE para entrenar, evaluar y registrar todo
    for params in ParameterGrid(param_grid):
        # AGREGADO: n_jobs=-1 utiliza toda la potencia del procesador en paralelo
        model = RandomForestClassifier(**params, random_state=seed, n_jobs=-1)
        model.fit(X_train_embeddings, y_train)

        # Predicciones
        y_pred_train = model.predict(X_train_embeddings)
        y_pred_val = model.predict(X_val_embeddings)

        # Métricas
        score_train = accuracy_score(y_train, y_pred_train)
        score_val = accuracy_score(y_val, y_pred_val)

        # Guardar historial para el reporte final
        historial_resultados.append((params, score_train, score_val))

        # Monitorear el mejor modelo
        if score_val > best_score:
            best_score = score_val
            best_params = params
            best_model = model

    # Imprimir los mejores resultados globales
    print("\n" + "="*50)
    print("¡BÚSQUEDA DE HIPERPARÁMETROS FINALIZADA!")
    print("="*50)
    print(f"Mejores hiperparámetros encontrados: {best_params}")

    # Evaluamos la precisión del mejor modelo final
    best_train_acc = accuracy_score(y_train, best_model.predict(X_train_embeddings))
    print(f"Accuracy en el conjunto de VALIDACIÓN: {best_score:.4f}")
    print(f"Accuracy en el conjunto de ENTRENAMIENTO: {best_train_acc:.4f}\n")

    # Reporte detallado usando el historial guardado en memoria (YA NO ENTRENA NADA AQUÍ)
    print("--- Historial Detallado de Combinaciones ---")
    for params, score_t, score_v in historial_resultados:
        print(f"Parámetros: {params} | Accuracy Train: {score_t:.4f} | Accuracy Val: {score_v:.4f}")

    return best_params, best_score, best_model
```

```python
def log_regression(X_t, y_t, X_v, y_v, seed=42):
    """
    Entrena un modelo de regresión logística y evalúa su desempeño en el conjunto de validación.
    Usamos parameter grid search para encontrar los mejores hiperparámetros y reportamos el accuracy obtenido en el conjunto de validación.
    return best_params, best_score, best_model
    """
    param_grid = {
        'C': [1, 10],
        'solver': ['liblinear']
    }
    best_score = 0
    best_params = None
    best_model = None
    for params in ParameterGrid(param_grid):
        model = LogisticRegression(**params, random_state=seed)
        model.fit(X_t, y_t)
        y_pred = model.predict(X_v)
        score = accuracy_score(y_v, y_pred)
        if score > best_score:
            best_score = score
            best_params = params
            best_model = model
    print(f"Mejores hiperparámetros: {best_params}")
    print(f"Accuracy en el conjunto de validación: {best_score:.4f}")
    print(f"Accuracy en el conjunto de entrenamiento: {accuracy_score(y_t, best_model.predict(X_t)):.4f}")
    return best_params, best_score, best_model
```

```python
def random_forest(X_t, y_t, X_v, y_v, seed=42):
    """
    Entrena un modelo de Random Forest y evalua su desempeno en el conjunto de validacion.
    Usamos parameter grid search para encontrar los mejores hiperparametros y reportamos el accuracy obtenido en el conjunto de validacion.
    return best_params, best_score, model
    """
    param_grid = {
        'n_estimators': [100, 200],
        'max_depth': [14],
        'min_samples_split': [5, 20],
        'min_samples_leaf': [4],
        'max_features': ['sqrt', 'log2']
    }

    best_score = 0
    best_params = None
    best_model = None

    # Lista para guardar los resultados del historial en la primera pasada
    historial_resultados = []

    print("Iniciando Grid Search optimizado de Random Forest...")

    # UN SOLO BUCLE: Hace todo el trabajo pesado una sola vez
    for params in ParameterGrid(param_grid):
        # CORRECCIÓN: Se agrega n_jobs=-1 para paralelizar el entrenamiento en la CPU
        model = RandomForestClassifier(**params, random_state=seed, n_jobs=-1)
        model.fit(X_t, y_t)

        # Predicciones necesarias
        y_pred_train = model.predict(X_t)
        y_pred_val = model.predict(X_v)

        # Cálculo de métricas
        score_train = accuracy_score(y_t, y_pred_train)
        score_val = accuracy_score(y_v, y_pred_val)

        # Guardamos en memoria para el reporte final sin volver a entrenar
        historial_resultados.append((params, score_train, score_val))

        if score_val > best_score:
            best_score = score_val
            best_params = params
            best_model = model

    # Reporte de los mejores resultados encontrados
    print("\n" + "="*50)
    print("¡BÚSQUEDA DE HIPERPARÁMETROS FINALIZADA!")
    print("="*50)
    print(f"Mejores hiperparametros: {best_params}")
    print(f"Accuracy en el conjunto de validacion: {best_score:.4f}")

    best_train_acc = accuracy_score(y_t, best_model.predict(X_t))
    print(f"Accuracy en el conjunto de entrenamiento: {best_train_acc:.4f}\n")

    # Reporte detallado usando la información recolectada (Cero costo de cómputo aquí)
    print("--- Historial Detallado de Combinaciones ---")
    for params, score_t, score_v in historial_resultados:
        print(f"Parametros: {params}, Accuracy Train: {score_t:.4f}, Accuracy Val: {score_v:.4f}")

    return best_params, best_score, best_model
```
