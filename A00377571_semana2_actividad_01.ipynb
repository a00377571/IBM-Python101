*   **Pregunta 1.**

Busca y elimina todos los saltos de línea '\n' que se encuentran al final de cada comentario.

Una vez finalizado, imprime los primeros 10 comentarios del resultado obtenido.


```python
docs_SinSalto = [re.sub(r'\n$', '', doc) for doc in docs]
```

*   **Pregunta 2.**  

Busca e imprime todas las palabras que terminan con dos o más signos de admiración seguidos, por ejemplo "!!!".

Debes imprimir tanto la palabra como la totalidad de signos de admiración que le siguen.

Indica cuántos resultados obtuviste.



```python
palabras_admiracion = []
regex = r'\b(\w+)(!{2,})'

for doc in docs_SinSalto:
    matches = re.findall(regex, doc)
    for match in matches:
        palabras_admiracion.append(match)

for palabra, signoadmiracion in palabras_admiracion:
    print(f"Palabra: {palabra}, Signos de admiración: {signoadmiracion}")

```

```python
print(f"\nTotal palabras con admiración obtenidas: {len(palabras_admiracion)}")
```

*   **Pregunta 3.**  

Busca e imprime todas las palabras que están escritas totalmente en mayúsculas. Cada coincidencia debe ser una sola palabra.

Indica cuántas palabras encontraste.



```python
mayusculas = []
regex = r'\b[A-Z]+\b'

for doc in docs_SinSalto:
    busqueda = re.findall(regex, doc)
    for palabra in busqueda:
        mayusculas.append(palabra)

for palabra in mayusculas:
    print(f"Palabra en mayúsculas: {palabra}")
```

```python
print(f"\n Total palabras en mayúsculas: {len(mayusculas)}")
```

*   **Pregunta 4.**  

Busca e imprime los comentarios en donde todos los caracteres alfabéticos (letras) están en mayúsculas.

Cada coincidencia encontrada debe ser todo el comentario/enunciado.

Indica cuántos resultados obtuviste.


```python
comentarios_mayusculas = []

for comentario in docs_SinSalto:
    # Valirar que  no hay minusculas y que hay almenos una letra mayúscula
    if not re.search(r'[a-z]', comentario) and re.search(r'[A-Z]', comentario):
        comentarios_mayusculas.append(comentario)

for comentario in comentarios_mayusculas:
    print(f"Comentario en mayúsculas: {comentario}")

```

```python
print(f"\nTotal de comentarios en mayúsculas: {len(comentarios_mayusculas)}")
```

*   **Pregunta 5.**  

Busca e imprime todas las palabras que tengan una vocal acentuada, del tipo á, é, í, ó, ú.

Indica cuántos resultados obtuviste.

```python
vocales_acentuadas = []
# metodo Regex para encontrar palabras que contienen alguna vocál acentuada.
regex = r'\b\w*[áéíóúÁÉÍÓÚ]\w*\b'

for doc in docs_SinSalto:
    coincidencia = re.findall(regex, doc)
    for word in coincidencia:
        if word not in vocales_acentuadas:
            vocales_acentuadas.append(word)


for word in vocales_acentuadas:
    print(f"Palabra con vocal acentuada: {word}")
```

```python
print(f"\nTotal de palabras con vocales acentuadas encontradas: {len(vocales_acentuadas)}")
```

*   **Pregunta 6.**  

Busca e imprime todas las cantidades numéricas monetarias, enteras o con decimales, que inician con el símbolo $\$$.

Indica cuántos resultados obtuviste.

```python
cantidades_monetarias = []
# Método Regex para encontrar el signo de '$' seguido de uno o más digitos, opcionalmente seguido por un punto decimal o mas digitos.
regex = r'\$\d+(?:\.\d+)?'

for doc in docs_SinSalto:
    match = re.findall(regex, doc)
    for conjunto in match:
        cantidades_monetarias.append(conjunto)

for conjunto in cantidades_monetarias:
    print(f"Cantidad monetaria: {conjunto}")
```

```python
print(f"\nTotal de cantidades monetarias encontradas: {len(cantidades_monetarias)}")
```

*   **Pregunta 7.**  

Busca e imprime todas las palabras que sean variantes de la palabra "love", sin importar si incluyen mayúsculas o minúsculas, o la manera en que esté conjugada o alguna otra variación que se haga con dicha palabra.

Indica cuántos resultados obtuviste.

```python
love_variants = []
# Regex to find 'love' and its common variations like loved, loving, loves
# \b ensures whole word match, (?:...)? makes the group optional for variations
regex = r'\blov(?:e|ed|ing|es)?\b'

for doc in docs_SinSalto:
    matches = re.findall(regex, doc, re.IGNORECASE) # re.IGNORECASE for case-insensitive matching
    for variant in matches:
        love_variants.append(variant)

for variant in love_variants:
    print(f"Variante de 'love': {variant}")
```

```python
print(f"\nTotal de variantes de 'love' encontradas: {len(love_variants)}")
```

*   **Pregunta 8.**  

Busca e imprime todas las palabras, variantes de "so" y "good", que tengan dos o más "o" en "so" y 3 o más "o" en good.

Indica cuántas encontraste.


```python
so_good_variants = []
# Regex to find 'so' variants with 2+ 'o's or 'good' variants with 3+ 'o's
# \b ensures whole word match
regex = r'\b(s[oO]{2,}|g[oO]{3,}d)\b'

for doc in docs_SinSalto:
    matches = re.findall(regex, doc, re.IGNORECASE) # re.IGNORECASE for case-insensitive matching
    for variant in matches:
        so_good_variants.append(variant)

for variant in so_good_variants:
    print(f"Variante encontrada: {variant}")
```

```python
print(f"\nTotal de variantes de 'so'/'good' encontradas: {len(so_good_variants)}")
```

*   **Pregunta 9.**  

Busca e imprime todas las palabras que tengan una longitud mayor estrictamente a 10 caracteres alfabéticos.

No se consideran los signos de puntuación o caracteres especiales en la longitud de estas cadenas, solo caracteres alfabéticos en mayúsculas o minúsculas.

Indica la cantidad de palabras encontradas.


```python
long_words = []
# Regex to find words consisting only of alphabetic characters (case-insensitive)
regex = r'\b[a-zA-Z]+\b'

for doc in docs_SinSalto:
    # Find all alphabetic words in the document
    alphabetic_words = re.findall(regex, doc)
    for word in alphabetic_words:
        # Check if the length of the word is strictly greater than 10
        if len(word) > 10:
            long_words.append(word)

for word in long_words:
    print(f"Palabra larga: {word}")
```

```python
print(f"\nTotal de palabras con longitud mayor a 10: {len(long_words)}")
```

*   **Pregunta 10.**  

Busca e imprime todas las palabras que inician con una letra mayúscula y terminan con una minúscula, pero que además no sea la primera palabra del comentario/string.

Indica la cantidad de resultados obtenidos.

```python
words_capital_start_lowercase_end = []
regex = r'\b[A-Z][a-zA-Z]*[a-z]\b'

for doc in docs_SinSalto:
    # Tokenize the document into words using a simple split for this purpose
    words = re.findall(r'\b\w+\b', doc) # Find all words first to check their position
    for i, word in enumerate(words):
        if i == 0:
            continue # Skip the first word of the comment

        # Check if the word matches the regex pattern
        if re.match(regex, word):
            words_capital_start_lowercase_end.append(word)

for word in words_capital_start_lowercase_end:
    print(f"Palabra encontrada: {word}")
```

```python
print(f"\nTotal de palabras encontradas: {len(words_capital_start_lowercase_end)}")
```

*   **Pregunta 11.**  

Busca e imprime la secuencia de dos o más palabras que están separadas por un guion, "-", sin que tengan espacios en blanco entre ellas.

Por ejemplo "Go-Kart" sería válido, pero "Go  -Kart" o "Go  -  Kart" no lo serían.

Indica la cantidad de resultados obtenidos.

```python
hyphenated_words = []
# Regex to find two or more words connected by a hyphen without spaces
regex = r'\b\w+(?:-\w+)+\b'

for doc in docs_SinSalto:
    matches = re.findall(regex, doc)
    for match in matches:
        hyphenated_words.append(match)

for word_seq in hyphenated_words:
    print(f"Secuencia con guion: {word_seq}")
```

```python
print(f"\nTotal de secuencias con guion encontradas: {len(hyphenated_words)}")
```

*   **Pregunta 12.**  

Busca e imprime todas las palabras que terminan en "ing" o "ed".

Indica la cantidad de palabras que encontraste de cada una.

```python
ending_ing = []
ending_ed = []

regex_ing = r'\b\w+ing\b'
regex_ed = r'\b\w+ed\b'

for doc in docs_SinSalto:
    matches_ing = re.findall(regex_ing, doc, re.IGNORECASE)
    for word in matches_ing:
        ending_ing.append(word)

    matches_ed = re.findall(regex_ed, doc, re.IGNORECASE)
    for word in matches_ed:
        ending_ed.append(word)

print("Palabras terminadas en 'ing':")
for word in ending_ing:
    print(word)

print("\nPalabras terminadas en 'ed':")
for word in ending_ed:
    print(word)
```

```python
print(f"\nTotal de palabras terminadas en 'ing': {len(ending_ing)}")
print(f"Total de palabras terminadas en 'ed': {len(ending_ed)}")
```

#**Parte 3. Proceso de limpieza.**

*   **Pregunta 13.**  

Ahora realiza un proceso de limpieza del corpus que incluya los siguientes procesos:

*   Solo se deben considerar caracteres alfabéticos. Es decir, se eliminan todos los signos de puntuación y caracteres especiales.
*   Todos los caracteres alfabéticos se transforman a minúsculas.
*   Se deben eliminar todos los espacios en blanco adicionales que se puedan encontrar en cada comentario.

Al finalizar dicho proceso de limpieza, imprime el resultado de los primeros 10 comentarios resultantes.
   




```python
cleaned_docs = []

for doc in docs_SinSalto:
    # 1. Keep only alphabetic characters, replacing others with spaces
    doc_cleaned = re.sub(r'[^a-zA-Z]', ' ', doc)
    # 2. Transform all alphabetic characters to lowercase
    doc_cleaned = doc_cleaned.lower()
    # 3. Eliminate additional white spaces and strip leading/trailing spaces
    doc_cleaned = re.sub(r'\s+', ' ', doc_cleaned).strip()
    cleaned_docs.append(doc_cleaned)

```

```python
# Print the first 10 cleaned comments
print("Primeros 10 comentarios limpios:")
for i in range(10):
    print(cleaned_docs[i])
```

*   **Pregunta 14.**  

Con el resultado de la limpieza obtenido en la pregunta anterior, realiza ahora un proceso de tokenización por palabras del corpus.

Es decir, al final de este proceso de tokenización, debes tener como resultado una lista de listas, donde cada comentario estará tokenizado por palabras.

Al terminar calcula el total de tokens obtenido en todo el corpus.

```python
tokenized_docs = []
for doc in cleaned_docs:
    tokenized_docs.append(doc.split()) # Simple space-based tokenization

```

```python
# Calculate the total number of tokens
total_tokens = sum(len(comment_tokens) for comment_tokens in tokenized_docs)

print("Primeros 5 comentarios tokenizados:")
for i in range(min(5, len(tokenized_docs))):
    print(tokenized_docs[i])

print(f"\nTotal de tokens en todo el corpus: {total_tokens}")
```
